```python
import numpy as np
import pandas as pm 
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sb
sb.set()
```

```python
raw_data = pm.read_csv('Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1,'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat = 0.0017*data['SAT'] + 0.275
flg = plt.plot(data['SAT'],yhat_no,lw=2,c='#006837')
flg = plt.plot(data['SAT'],yhat_yes,lw=2,c='#a50026')
flg = plt.plot(data['SAT'],yhat,lw=3,c='#4C7280',label='regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
new_data = pm.DataFrame({'const':1,'SAT':[1700,1670],'Attendance':[0,1]})
new_data = new_data[['const','SAT','Attendance']]
new_data
```

```python
predictions = results.predict(new_data)
predictions
```
